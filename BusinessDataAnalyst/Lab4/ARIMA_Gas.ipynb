```python
# Các thư viện cần thiết
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from pmdarima.arima import auto_arima
from statsmodels.tsa.stattools import adfuller
```

```python
# Đọc file dữ liệu Gas & Electric
df = pd.read_csv('Gas&Electric.csv')
```

```python
# Vẽ biểu đồ chuỗi thời gian
plt.plot(df.index, df["Gas Use"])
plt.xlabel("Index")
plt.ylabel("Gas Use")
plt.show()
```

```python
# Chia train test theo tỉ lệ 7:3
df = df[['Gas Use']]
df = df.dropna() # Loại bỏ các hàng trống
df = df.reset_index(drop=True) # Thiết lập lại chỉ mục

# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
train_size = int(0.7 * len(df))
test_size = len(df) - train_size

train_data = df[:train_size]
test_data = df[train_size:]
```

```python
# Quá trình Training
x_train = np.array(train_data.index).reshape(-1, 1)  #Lấy chỉ mục Index
y_train = np.array(train_data['Gas Use']) #Lấy dữ liệu Gas Use

# Tìm kiếm các tham số tốt nhất bằng auto_arima
model = auto_arima(y_train, start_p=1, start_q=1,
                   max_p=1, max_q=1, m=12,  # m là chu kỳ mùa vụ, ở đây là mỗi năm
                   start_P=0, seasonal=True,
                   d=1, D=1, trace=True,
                   error_action='ignore',
                   suppress_warnings=True,
                   stepwise=True)

# Fit mô hình với các tham số tốt nhất được tìm kiếm
model.fit(y_train)
```

```python
# Quá trình testing 
x_test = np.array(test_data.index).reshape(-1, 1) #Lấy chỉ mục Index
y_test = np.array(test_data['Gas Use']) #lấy dữ liệu Gas Use

y_pred = model.predict(n_periods=len(x_test))
```

```python
# Quá trình tạo index predict 12 month tiếp theo

# Tạo ra mảng chỉ mục của x_next_12_months
x_next_12_months = np.array(range(df.index[-1] + 1, df.index[-1] + 13)).reshape(-1, 1)

# Tạo ra mảng dữ liệu Gas Use dự đoán của y_next_12_months
y_next_12_months = model.predict(n_periods=len(x_next_12_months))

# In ra dữ liệu Gas Use dự đoán trong 12 tháng tiếp theo:
print('Lượng Gas dự đoán trong 12 tháng tới:')
print(y_next_12_months)
```

```python
# Tính MAE - Mean Absolute Error - Sai số trung bình tuyệt đối
test_mae = np.mean(np.abs(y_pred - y_test))
print("Testing MAE:", test_mae)

# Tính RMSE - Root Mean Squared Error - Sai số bình phương trung bình
test_rmse = np.sqrt(np.mean((y_pred - y_test)**2))
print('Testing RMSE:', test_rmse)

# Tính MAPE - Mean Absolute Percentage Error - Tỷ lệ phần trăm sai số tuyệt đối trung bình
test_mape = np.mean(np.abs(y_pred - y_test)/np.abs(y_test))
print('Testing MAPE:', test_mape)
```

```python
# Vẽ hình
plt.plot(x_train, y_train)
plt.plot(x_test, y_test)
plt.plot(x_test, y_pred)
plt.plot(x_next_12_months, y_next_12_months)
plt.legend(['Train', 'Test', 'Predictions','Next12Months'])
plt.show()
```
