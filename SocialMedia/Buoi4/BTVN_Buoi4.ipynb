```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
DG = nx.Graph()
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D'), ('A', 'F'), ('B', 'D'), ('B', 'E'), ('B', 'G'), ('C', 'F'), ('C', 'D'),
                   ('D', 'F'), ('D', 'G'), ('D', 'E'), ('E', 'G'), ('F', 'H'), ('F', 'G'), ('G', 'H'), ('H', 'I'), ('I', 'J')])
```

```python
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0),node_size=600,cmap='coolwarm',font_size='14',font_color='white')
```

```python
print(f'Betweenness centrality: {nx.betweenness_centrality(DG)}')
```

```python
print(f'Closeness centrality: {nx.closeness_centrality(DG)}')
```
