```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
DG = nx.DiGraph()
DG.add_edges_from([('v1', 'v2'), ('v1', 'v3'), ('v3', 'v2'), ('v3', 'v4'), ('v2', 'v4'), ('v4', 'v5'), ('v4', 'v6')])
```

```python
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0),node_size=600,cmap='coolwarm',font_size='14',font_color='white')
```

```python
print(f'Degree centrality: {nx.degree_centrality(DG)}')
```

```python
print(f'Between centrality: {nx.betweenness_centrality(DG, None, False, None, False, None)}')
```

```python
print(f'Closeness centrality: {nx.closeness_centrality(DG)}')
```
