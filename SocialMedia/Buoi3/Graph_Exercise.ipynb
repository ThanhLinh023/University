```python
import networkx as nx
import matplotlib.pyplot as plt
#import numpy as np
```

```python
DG = nx.Graph()
DG.add_edges_from([('A', 'B'),('B', 'C'),('B', 'D'),('C', 'E'),('E', 'D'),('E', 'H'),('E', 'G'),('E', 'F'),('H', 'G'),('D', 'G')])
```

```python
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0), arrows=False,node_size=600,cmap='coolwarm',font_size='14',font_color='white')
```

```python
node_cluster = nx.clustering(G=DG)
print('Node Clustering')
for node, cl_coeff in node_cluster.items():
    print(f'Node {node}: {cl_coeff}')
    nx.draw_networkx(DG, with_labels=True, node_color='green')
```
