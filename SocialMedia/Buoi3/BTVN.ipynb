```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
DG = nx.Graph()
DG.add_edges_from([('0', '1'),('0', '3'),('0', '4'),('1', '2'),('1', '5'),('1', '4'),('2', '5'),('2', '6'),('3', '4'),('3', '7'),
                   ('4', '5'),('4', '7'),('4', '8'),('5', '6'),('5', '8'),('5', '9'),('6', '9'),('7', '8'),('8', '9')])
```

```python
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0), arrows=False,node_size=600,cmap='coolwarm',font_size='14',font_color='white')
```

```python
for node, degree in DG.degree():
    print(f"Node {node} - Bậc {degree}")
```

```python
degree_sequence = [degree for node, degree in DG.degree()]
degree_distribution = {deg: degree_sequence.count(deg) for deg in set(degree_sequence)}

print("\nPhân phối bậc của đỉnh:")
for degree, count in degree_distribution.items():
    print(f"Bậc {degree}: {count} / {len(DG.nodes())}")
```

```python
node_cluster = nx.clustering(G=DG)
print('Clustering Coefficient')
for node, cl_coeff in node_cluster.items():
    print(f'Node {node}: {cl_coeff}')
    nx.draw_networkx(DG, with_labels=True, node_color='yellow')
```
