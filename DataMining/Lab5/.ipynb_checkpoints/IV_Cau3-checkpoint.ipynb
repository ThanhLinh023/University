```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
```



> **1. Đọc dữ liệu**


```python
df = pd.read_csv('/content/drive/MyDrive/Data Mining/Lab5/Mall_Customers.csv')
```

**2. Dùng Histogram để biểu diễn phân phối**

```python
plt.style.use('fivethirtyeight')
plt.figure(1, figsize=(15,6))
n = 0
for x in ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']:
  n += 1
  plt.subplot(1, 3, n)
  plt.subplots_adjust(hspace=0.5, wspace=0.5)
  sns.distplot(df[x], bins = 20)
  plt.title('Distplot of {}'.format(x))
plt.show()
```

**3. Thống kế số mẫu dữ liệu theo giới tính**

```python
plt.figure(1, figsize=(15,5))
sns.countplot(y='Gender', data=df)
plt.show()
```

**4. Phân bố của thuộc tính Tuổi và Thu nhập hằng năm theo giới tính**

```python
plt.figure(1, figsize=(15,6))
for g in ['Male', 'Female']:
  plt.scatter(x = 'Age', y = 'Annual Income (k$)', data = df[df['Gender'] == g],
              s = 200, alpha = 0.5, label = g)
plt.xlabel('Age'), plt.ylabel('Annual Income (k$)')
plt.title('Age vs Annual Income w.r.t Gender')
plt.legend()
plt.show()
```

**5. Phân bố của thuộc tính Thu nhập hằng năm và Số điểm mua sắm theo Giới tính**

```python
plt.figure(1, figsize=(15,6))
for g in ['Male', 'Female']:
  plt.scatter(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', data = df[df['Gender'] == g],
              s = 200, alpha = 0.5, label = g)
plt.xlabel('Annual Income'), plt.ylabel('Spending Score')
plt.title('Annual Income vs Spending Score w.r.t Gender')
plt.legend()
plt.show()
```

**6. Sử dụng ELBOW để xác định số cụm cần thiết**

```python
X1 = df[['Age', 'Spending Score (1-100)']].iloc[: ,:].values
iner = []
for n in range(1, 11):
  algo = KMeans(n_clusters = n, init='random')
  algo.fit(X1)
  iner.append(algo.inertia_)
plt.figure(1, figsize=(15, 6))
plt.plot(np.arange(1, 11), iner, 'o')
plt.plot(np.arange(1, 11), iner, '-', alpha = 0.5)
plt.xlabel('Number of Clusters'), plt.ylabel('Inertia')
plt.show()
```

```python
algo = KMeans(n_clusters = 4, init='random')
algo.fit(X1)
labels1 = algo.labels_
centroids1 = algo.cluster_centers_
```

**7. Biểu diễn kết quả bằng biểu đồ**

```python
plt.figure(1, figsize=(15,7))
plt.clf()
plt.scatter(x = 'Age', y = 'Spending Score (1-100)', data = df, c = labels1, s = 200)
plt.scatter(x = centroids1[: , 0], y = centroids1[: , 1], s = 300, c='red', alpha=0.5)
plt.ylabel('Spending Score (1-100)'), plt.xlabel('Age')
plt.show()
```

**8. Đánh giá kết quả gom cụm bằng hệ số Dáng điệu**

```python
metrics.silhouette_score(X1, labels1, metric='euclidean')
```

**9. Thực hiện gom cụm với các trường hợp khác**

```python
# Elbow to define k: Annual Income and Spending Score
X2 = df[['Annual Income (k$)', 'Spending Score (1-100)']].iloc[: ,:].values
iner = []
for n in range(1, 11):
  algo = KMeans(n_clusters = n, init='random')
  algo.fit(X2)
  iner.append(algo.inertia_)
plt.figure(1, figsize=(15, 6))
plt.plot(np.arange(1, 11), iner, 'o')
plt.plot(np.arange(1, 11), iner, '-', alpha = 0.5)
plt.xlabel('Number of Clusters'), plt.ylabel('Inertia')
plt.show()
```

```python
# Elbow to define k: Age, Annual Income and Spending Score
X3 = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].iloc[: ,:].values
iner = []
for n in range(1, 11):
  algo = KMeans(n_clusters = n, init='random')
  algo.fit(X3)
  iner.append(algo.inertia_)
plt.figure(1, figsize=(15, 6))
plt.plot(np.arange(1, 11), iner, 'o')
plt.plot(np.arange(1, 11), iner, '-', alpha = 0.5)
plt.xlabel('Number of Clusters'), plt.ylabel('Inertia')
plt.show()
```

```python
algo2 = KMeans(n_clusters = 5, init='random')
algo2.fit(X2)
labels2 = algo2.labels_
centroids2 = algo2.cluster_centers_

plt.figure(1, figsize=(15,7))
plt.clf()
plt.scatter(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', data = df, c = labels2, s = 200)
plt.scatter(x = centroids2[: , 0], y = centroids2[: , 1], s = 300, c='red', alpha=0.5)
plt.ylabel('Spending Score (1-100)'), plt.xlabel('Annual Income (k$)')
plt.show()
```

```python
algo3 = KMeans(n_clusters = 5, init='random')
algo3.fit(X3)
labels3 = algo3.labels_
centroids3 = algo3.cluster_centers_

df['label3'] = labels3
trace1 = go.Scatter3d(
    y=df['Spending Score (1-100)'],
    x=df['Age'],
    z=df['Annual Income (k$)'],
    mode='markers',
    marker=dict(
        color=df['label3'],
        size=20,
        line=dict(color=df['label3'], width=12),
        opacity=0.8
    )
)

dat = [trace1]

layout = go.Layout(
    title='Clusters',
    scene=dict(
        xaxis=dict(title='Age'),
        yaxis=dict(title='Spending Score'),
        zaxis=dict(title='Annual Income')
    )
)

fig = go.Figure(data=dat, layout=layout)

py.offline.iplot(fig)
```
