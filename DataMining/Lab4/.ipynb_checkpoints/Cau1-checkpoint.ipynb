```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris= load_iris()
```

```python
print('Feature names: ', iris.feature_names)
print('Target names: ', iris.target_names)
print('Data shape: ', iris.data.shape)
print('Target shape: ', iris.target.shape)

for i in range(5):
  print("Sample ", i + 1, ": ", iris.data[i], " - Label: ", iris.target[i])

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.describe()
```

```python
X = iris.data
Y = iris.target

X = X / X.max(axis=0)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

tree_pred = tree.predict(X_test)
tree_accuracy = accuracy_score(y_test, tree_pred)
print('Decision Tree Accuracy: ', tree_accuracy)

tree_report = classification_report(y_test, tree_pred)
print(tree_report)
tree_conf_matrix = confusion_matrix(y_test, tree_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(tree_conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
naive = GaussianNB()
naive.fit(X_train, y_train)

naive_pred = naive.predict(X_test)
naive_accuracy = accuracy_score(y_test, naive_pred)
print('Naive Bayes Accuracy: ', naive_accuracy)

naive_report = classification_report(y_test, naive_pred)
print(naive_report)
naive_conf_matrix = confusion_matrix(y_test, naive_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(naive_conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_pred)
print('KNN Accuracy: ', knn_accuracy)

knn_report = classification_report(y_test, knn_pred)
print(naive_report)
knn_conf_matrix = confusion_matrix(y_test, knn_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(knn_conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

neural_network_model = Sequential([
    Dense(10, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(20, activation='relu'),
    Dense(3, activation='softmax')
])

neural_network_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

neural_network_model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1)
```

```python
import numpy as np
neural_pred_prob = neural_network_model.predict(X_test)
neural_pred = np.argmax(neural_pred_prob, axis=1)

neural_accuracy = accuracy_score(y_test, neural_pred)
print('Neural Network Accuracy: ', neural_accuracy)

neural_report = classification_report(y_test, neural_pred)
print(neural_report)
neural_conf_matrix = confusion_matrix(y_test, neural_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(neural_conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
test_sample = np.array([[6.2,	2.9,	4.3,	1.3],
                        [5.1,	3.5,	1.4,	0.2],
                        [7.3,	2.8,	6.4,	2.1]])
test_sample = test_sample / test_sample.max(axis=0)

test_tree = tree.predict(test_sample)
test_naive = naive.predict(test_sample)
test_knn = knn.predict(test_sample)
test_neural_prob = neural_network_model.predict(test_sample)
test_neural = np.argmax(test_neural_prob, axis=1)

print('Decision Tree Predict: ', test_tree)
print('Naive Bayes Predict: ', test_naive)
print('KNN Predict: ', test_knn)
print('Neural Network Predict: ', test_neural)
```

**Nhận xét các mô hình**
- 3 mô hình Decision Tree, Naive Bayes và K Nearest Neighbor có kết quả và độ chính xác tương tự nhau (accuracy = 1 và f1-score = 1)
- Mô hình Neural Network có sự khác biệt về độ chính xác với 3 mô hình trên (accuracy = 0.966667 và f1-score = 1)
- Khi dự đoán thì Naive Bayes khác so với 3 mô hình còn lại
